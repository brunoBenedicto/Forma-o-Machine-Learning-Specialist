```python
from PIL import Image
import numpy as np
```

```python
def converter_para_escala_de_cinza(img_color):
    largura, altura = img_color.size
    img_cinza = Image.new('L', (largura, altura))

    for x in range(largura):
        for y in range(altura):
            r, g, b = img_color.getpixel((x, y))
            tom_cinza = int(0.2989 * r + 0.587 * g + 0.114 * b)
            img_cinza.putpixel((x, y), tom_cinza)
    return img_cinza 
```

```python
def converter_para_preto_e_branco(imagem):
    largura, altura = imagem.size
    img_preto_e_branco = imagem.copy()  # Cria uma c√≥pia da imagem original

    for x in range(largura):
        for y in range(altura):
            r, g, b = imagem.getpixel((x, y))
            tom_cinza = int(0.2989 * r + 0.587 * g + 0.114 * b)

            if tom_cinza < 128:
                img_preto_e_branco.putpixel((x, y), (0, 0, 0))  # Define o pixel como preto
            else:
                img_preto_e_branco.putpixel((x, y), (255, 255, 255))  # Define o pixel como branco

    return img_preto_e_branco
```

```python

arquivo = "/content/cores-768x511.jpg" 
img_color = Image.open(arquivo)
img_cinza = converter_para_escala_de_cinza(img_color)
img_preto_e_branco = converter_para_preto_e_branco(img_color)
img_color.show()
img_cinza.show()
img_preto_e_branco.show()
```
